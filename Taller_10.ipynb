```python
import numpy as np
import pandas as pd
import seaborn as sb
import os
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/adiacla/bigdata/master/titanic.csv')
data
```

```python
df = data.copy(deep=True)
df = data.drop('Cabin', axis=1)
df = df.dropna()
df
```

```python
mapa1 = { 'male': 1, 'female': 2}
df['Sex'] = df['Sex'].map(mapa1)

mapa2 = { 'S': 0, 'C': 1, 'Q': 2}
df['Embarked'] = df['Embarked'].map(mapa2)

df
```

```python
df = df.drop('Ticket', axis=1)
df = df.drop('Name', axis=1)
df = df.drop('PassengerId', axis=1)
df
```

```python
#Seleccionar variables
pedictores = df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
objetivo = df.Survived
```

```python
#Partición validación cruzada
pred_train, pred_test, res_train, res_test = train_test_split(pedictores, objetivo, test_size=0.2)
```

```python
#Modelo SVM
SVM_Model = SVC(gamma='auto', kernel='rbf', C=0.5)
SVM_Model.fit(pred_train, res_train)
```

```python
res_test
```

```python
print (f'Precisión: {SVM_Model.score(pred_test, res_test):.4f}')
```
